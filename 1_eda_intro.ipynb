```python
#importing libraries
import ast
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading dataset
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset['train'].to_pandas()
```

```python
#cleaning up the data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df
```

```python
df['job_skills'].apply(type).value_counts()
```

# Filter for US Data Analyst Roles

```python
df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short']=='Data Analyst')]
df_DA_US
```

```python
df_plot = df_DA_US['job_location'].value_counts().head(10).reset_index()
df_plot.columns = ['job_location', 'count']  # Rename columns properly
```

```python
sns.barplot(data=df_plot, x='count', y='job_location')
sns.despine()
plt.title('Counts of job locations for Data Analyst in the US')
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```

```python
dict_column = {'job_work_from_home':'Work from Home Offered',
              'job_no_degree_mention': 'Degree Requirement',
              'job_health_insurance': 'Health Insurance Offered'}

fig,ax = plt.subplots(1,3)
fig.set_size_inches(12,5)

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), labels = ['False', 'True'], autopct = '%1.1f%%',startangle = 90)
    ax[i].set_title(title)
    
plt.show()
```

```python
df_plot = df_DA_US['company_name'].value_counts().head(10).reset_index()
df_plot.columns = ['company_name', 'count']  # Rename columns properly
sns.barplot(data=df_plot, x='count', y='company_name')
sns.despine()
plt.title('Counts of Companies for Data Analyst in the US')
plt.xlabel("Number of Jobs")
plt.ylabel("")
plt.show()
```
